```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
y=np.random.randn(100)
print(y)
```

```python
x=[x for x in range(len(y))]
print(x)
```

```python
plt.plot(x,y,'-')
plt.fill_between(x,y,200,where=(y>195),facecolor='g',alpha=0.6)

plt.title("Gr√°fico ejemplo")
plt.show()
```
