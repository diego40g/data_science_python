```python
import pandas as pd
from numpy import random
```

```python
mean = 10
standard_deviation = 3
print(mean)
print(standard_deviation)
```

```python
data = random.normal(loc = mean, 
                     scale = standard_deviation, 
                     size = (20, 5))
print(data)
```

```python
df = pd.DataFrame(data, 
                  columns = list('ABCDE'), 
                  index = range(1, data.shape[0]+1))
print(df)
```
