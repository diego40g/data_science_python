```python
import numpy as np
```

```python
x=np.random.rand(100)
print(x)
```

```python
y=np.random.uniform(size=100)
print(y)
```

```python
print(f'Correlacion matrices: {np.corrcoef(x,y)}\n')
```

```python
print(f'Covarainza de las matrices: {np.cov(x,y)}\n')
```
