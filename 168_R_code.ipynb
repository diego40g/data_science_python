```r
library(titanic)
titanic_train$Survived <- factor(titanic_train$Survived);
box1<-ggplot(titanic_train, aes(Survived, Age, fill=Survived))+
  geom_violin()+ geom_boxplot(fill="black", color="white",
               lwd=1.1, width=0.1)+
   scale_fill_brewer(palette = "Set1")+ theme(legend.position = "none")+
  xlab("Supervivencia")+ ylab("Edad (aos)")+
  scale_x_discrete(breaks=c("0","1"), labels=c("No", "S"));
box1
```
