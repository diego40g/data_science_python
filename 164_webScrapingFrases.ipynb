```python
import socket

Server = socket.socket(socket.AF_INET,socket.SOCK_STREAM)
Server.connect(('www.w3.org', 80))
Read = 'GET https://www.w3.org/Summary.txt HTTP/1.0\r\n\r\n'.encode()
Server.send(Read)
```

```python
while True:
    info = Server.recv(512)
    if len(info)<1:
        break
    print(info.decode(),end='')
Server.close()   
```

```python
import urllib.request

variable = urllib.request.urlopen('https://www.w3.org')

for x in variable:
    print(x.decode().strip())
```

```python
pagina_web=input('Enter - ')
codigo=urllib.request.urlopen(pagina_web)
bea=BeautifulSoup(codigo)

etiqueta=bea('a')
for x in etiqueta:
    print('Etiqueta "a" completa: ',x)
    print('Dirección: ', x.get('href'))
    print('Contenido: ',x.contents)
    print('Atributo: ',x.attrs)
    print('\n')
```
