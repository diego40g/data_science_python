```python
import numpy as np
```

```python
arr=np.random.normal(40,0.5,size=(300))
print(arr)
```

```python
print(f'Media: {np.mean(arr)}')
```

```python
print(f'Mediana: {np.median(arr)}')
```

```python
print(f'Desviacion estandar: {np.std(arr)}' )
```

```python
print(f'Varianza: {np.var(arr)}')
```
